```python

path_to_csv = '/content/drive/MyDrive/Colab Notebooks/breastcancer.csv'

import keras
import pandas
from tensorflow.keras.layers import Dense
from keras.models import Sequential
from keras.layers import Dense, Activation
from sklearn.preprocessing import StandardScaler, LabelEncoder

# load dataset
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

dataset = pd.read_csv(path_to_csv, header=None)

print(dataset.head())
print(dataset.info())

labelcode = LabelEncoder()

if dataset.iloc[:, -1].dtype == object:
  dataset.iloc[:, -1] = labelcode.fit_transform(dataset.iloc[:, -1])

for col in dataset.columns:
  if dataset[col].dtype == object:
    dataset[col] = labelcode.fit_transform(dataset[col])

dataset = dataset.values

X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:30], dataset[:,30],
                                                    test_size=0.25, random_state=87)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#independent = dataset.data
#dependent = dataset.target

#scale = sc.fit_transform(independent)
#print(scale)

np.random.seed(155)
my_first_nn = Sequential() # create model
my_first_nn.add(Dense(20, input_dim=30, activation='relu')) # hidden layer
my_first_nn.add(Dense(15, activation='relu'))
my_first_nn.add(Dense(10, activation='relu'))
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100,
                                     initial_epoch=0)
print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))
```
